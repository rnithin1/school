```
import numpy as np
import matplotlib.pyplot as plt
```

```
# Desired trajectory
t = np.linspace(0,10,1000)
xdes = np.sin(t)
ydes = np.zeros(len(t))
thtdes = np.zeros(len(t))
```

```
# ICs
x0 = 1
y0 = 0
tht0 = 0

error_x0 = x0 - xdes[0]
error_y0 = y0 - ydes[0]
error_tht0 = tht0 - thtdes[0]
```

```
# Control parameters
timeConst_x = 0.1
timeConst_y = 1
timeConst_tht = 1
```

```
# Closed-loop dynamics solution
error_x = np.exp(-t / timeConst_x)*error_x0
error_y = np.exp(-t / timeConst_y)*error_y0
error_tht = np.exp(-t / timeConst_tht)*error_tht0

x = error_x + xdes
y = error_y + ydes
tht = error_tht + thtdes

vx = np.diff(x)
vy = np.diff(y)
w = np.diff(tht)
```

```
# Plot the solution
plt.plot(t,x,t,xdes)
```
